```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_path = "./data/chars1.png"
```

```python
img = cv2.imread(img_path)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap="gray")
```

```python
canny_img = cv2.Canny(gray_img, 30, 150, 3)
plt.imshow(canny_img, cmap="gray")
```

```python
dilated_img = cv2.dilate(canny_img, (1, 1), iterations=0) 
plt.imshow(dilated_img, cmap="gray")
```

```python
(cnt, hierarchy) = cv2.findContours(dilated_img.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE) 
```

```python
print(f"Amount of characters: {len(cnt)}")
```

```python
height = []
width = []
for c in cnt:
    _, _, h, w = cv2.boundingRect(c)
    height.append(h)
    width.append(w)
height = np.array(height)
width = np.array(width)
```

```python
median_h = np.median(height)
median_w = np.median(width)
```

```python
print(f"median height = {median_h}, median width = {median_w}")
```
